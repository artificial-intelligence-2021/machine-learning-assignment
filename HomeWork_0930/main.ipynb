```python
from sql_helper import MySQL

user = input("请输入连接用户(aura+手机号码): ")
password = input("请输入连接密码: ")

mysql = MySQL()
drop_table_sql = """
drop table if exists iris;
"""
create_table_sql = """
create table if not exists iris(
    feature1 float,
    feature2 float,
    feature3 float,
    label    int
);    
"""
# 读取csv文件
with open(file="iris.csv", mode="r", encoding="utf8") as f:
    f.readline()
    data_values = ''
    while True:
        line = f.readline()
        if line:
            data = [float(ele) for ele in line.strip().split(",")]
            data_values = data_values + f'({data[0]},{data[1]},{data[2]},{data[3]}),'
        else:
            break

insert_sql = 'insert into iris values' + data_values
insert_sql = insert_sql[:-1]
read_iris_sql = """
    select * from iris;
"""
try:
    mysql.set_conn_info(user, password)
    mysql.db_con()
    print('数据已连接...')
    mysql.cursor.execute(query=drop_table_sql)
    mysql.cursor.execute(query=create_table_sql)
    print('成功建立iris表...')
    mysql.cursor.execute(query=insert_sql)
    print('数据已写入iris表...')
    table_cnt = mysql.cursor.execute(query=read_iris_sql)
    print(f'iris表有{table_cnt}条记录')
    
except Exception as ex:
    print(ex)
```
